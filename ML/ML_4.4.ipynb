```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

print("Индексы обучающей выборки:", X_train)
print("Индексы валидационной выборки:", X_val)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42, shuffle=True)

print("Индексы обучающей выборки:", X_train)
print("Индексы валидационной выборки:", X_val)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

model = LogisticRegression()
model.fit(X_train, y_train)

print("Коэффициенты модели:", model.coef_)

y_pred = model.predict(X_val)

print("Accuracy:", accuracy_score(y_val, y_pred))
print("F1-score:", f1_score(y_val, y_pred, average='weighted'))
```

```python
X_train, X_val, y_train, y_val = train_test_split(iris.data, iris.target, test_size=0.25, random_state=42, shuffle=True)

model = LogisticRegression()
model.fit(X_train, y_train)

print("Коэффициенты модели:", model.coef_)

y_pred = model.predict(X_val)

print("Accuracy:", accuracy_score(y_val, y_pred))
print("F1-score:", f1_score(y_val, y_pred, average='weighted'))
```

```python
from sklearn.model_selection import cross_val_score

model = LogisticRegression()
scores = cross_val_score(model, iris.data, iris.target, cv=5)

print("Значения метрик для каждого блока:", scores)
print("Среднее значение метрик:", scores.mean())
```

```python
from sklearn.model_selection import cross_val_score

model = LogisticRegression()
scores = cross_val_score(model, iris.data, iris.target, cv=5)

print("Значения метрик для каждого блока:", scores)
print("Среднее значение метрик:", scores.mean())
```

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

model = LogisticRegression()
scores = cross_val_score(model, iris.data, iris.target, cv=skf)

print("Значения метрик для каждого блока:", scores)
print("Среднее значение метрик:", scores.mean())
```

```python
from sklearn.model_selection import LeaveOneOut

loo = LeaveOneOut()

model = LogisticRegression()
scores = cross_val_score(model, iris.data, iris.target, cv=loo)

print("Значения метрик для каждого блока:", scores)
print("Среднее значение метрик:", scores.mean())
```
